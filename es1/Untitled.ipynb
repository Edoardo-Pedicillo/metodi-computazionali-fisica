```python
x4= np.loadtxt("data/rapporto4.txt", usecols=(0,1,2,3),unpack='true')
x10= np.loadtxt("data/rapporto10.txt", usecols=(0,1,2,3),unpack='true')
x15= np.loadtxt("data/rapporto15.txt", usecols=(0,1,2,3),unpack='true')
#print(x15,y15)
fig=plt.figure(figsize=(16,7))

plt.subplot(231)
plt.plot(x4[0],x4[3], color="firebrick", label="lung 4 ngen 300")
plt.xlabel('L(testo)/L(chiave) ')
plt.ylabel('# lettere in comune con la soluzione')

plt.legend()
plt.grid(True)

plt.subplot(234)
plt.plot(x4[0] ,x4[2]-x4[1], color="firebrick", label="lung 4 ngen 300")
plt.xlabel('L(testo)/L(chiave) ')
plt.ylabel('fitness chiave-fitness massima')

plt.legend()
plt.grid(True)

plt.subplot(232)
plt.plot(x10[0] ,x10[3], color="darkblue", label="lung 10 ngen 300")
plt.xlabel('L(testo)/L(chiave) ')
plt.ylabel('# lettere in comune con la soluzione')
plt.legend()
plt.grid(True)

plt.subplot(235)
plt.plot(x10[0] ,x10[2]-x10[1], color="darkblue", label="lung 10 ngen 300")
plt.xlabel('L(testo)/L(chiave) ')
plt.ylabel('fitness chiave-fitness massima')
plt.legend()
plt.grid(True)

plt.subplot(233)
plt.plot(x15[0],x15[3], color="forestgreen", label="lung 15 ngen 300")
plt.xlabel('L(testo)/L(chiave) ')
plt.ylabel('# lettere in comune con la soluzione')

plt.legend()
plt.grid(True)

plt.subplot(236)
plt.plot(x15[0] ,x15[2]-x15[1], color="forestgreen", label="lung 15 ngen 300")
plt.xlabel('L(testo)/L(chiave) ')
plt.ylabel('fitness chiave-fitness massima')
plt.legend()
plt.grid(True)
```

- Se il rapporto L(testo)/L(chiave) è piccolo ci aspettiamo che si possa verificare casi in cui si trovano fitness più alte della soluzione perchè le parole sono molto di meno rispetto agli n-grams, quindi per caso un individuo riesce ad avere molti n-grams senza dover per forza avere delle parole. In un testo più lungo le parole iniziano ad avere un peso decisivo nella fitness e quindi le soluzioni  migliori saranno quelle che conterranno più parole note. 

- Nella soluzione più lunga si osserva una oscillazione del numero di lettere in comune, questo può essere spiegato con il fatto che più una chiave è lunga più è difficile azzeccare tutte le lettere in quanto anche se si sbaglia di una lettera il numero effettivo di parole ed n-grams che si sbaglia è ridotto
