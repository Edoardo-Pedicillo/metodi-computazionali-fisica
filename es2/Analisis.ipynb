```python
d = np.loadtxt("data/analisi_cycle_cross.txt", usecols=(0,1),unpack='true',max_rows=1000)
e = np.loadtxt("data/analisi_bo_cross.txt", usecols=(0,1),unpack='true',max_rows=1000)
f = np.loadtxt("data/analisi_mo.txt", usecols=(0,1),unpack='true',max_rows=1000)
g = np.loadtxt("data/analisi_pm.txt", usecols=(0,1),unpack='true',max_rows=1000)

#fig=plt.figure(figsize=(14,7))
fig,ax = plt.subplots(figsize=(14,7))

ax.plot(d[0],d[1] , color="firebrick",label="cycle crossover")
# set x-axis label
ax.set_xlabel("# generazioni",fontsize=14)
# set y-axis label
ax.set_ylabel("fitness", fontsize=14)

ax.plot(e[0],e[1] , color="darkblue",label="based ordered crossover")

ax.plot(f[0],f[1] , color="orange",label="modified ordered crossover")

ax.plot(g[0],g[1] , color="purple",label="parially mapped crossover")


plt.axhline(y=930, color='lime', linestyle='dashed',label="fitness soluzione")
plt.grid(True)
plt.legend(fontsize='x-large')
```
