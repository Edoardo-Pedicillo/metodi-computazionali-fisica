```python
alphabet=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z',' ']

def txt_set(message):
    """
       elimina dal testo tutto ciÃ² che non sono caratteri e rende tutte le lettere minuscole.
       Restituisce una lista
    """
    message_list=list(message.lower())
    new_message_list=[]
   
    for i in message_list:
        
        if i in alphabet:
            new_message_list.append(i)

    return  "".join(new_message_list)

def get_fitness(self,message):
        
        fitness=0
        new_message=crittografia.decritt_permutation(message,"".join(self))
       
       
        frequency=[new_message.count(i) for i in alphabet]
       
        for i,item in enumerate(letters):
            
            fitness+=new_message.count(item)*2
            
        for i in digraphs:
            fitness+=new_message.count(i)*10
        
        for i in trigraphs:
            fitness+=new_message.count(i)*10
            
        for i in one_words:
            fitness+=new_message.count(" "+i+" ")*10
        
        for i in two_words:
            fitness+=new_message.count(' %s '%i)*20
        
        for i in three_words:
            fitness+=new_message.count(' %s '%i)*20
        
        for i in four_words:
            fitness+=new_message.count(' %s '%i)*50
        #return frequency
        
        for i, item in enumerate (new_message):
            if item==' ':
                try:
                    end=new_message.index(' ',i)
                    lenght_word=end-i
                    if lenght_word>7:
                        fitness-=10
                    
                except ValueError:
                    return fitness
        return fitness
```
